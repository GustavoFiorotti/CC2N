## Definição de **Estruturas de Repetição de Dados**:

$\longmapsto$ São usadas justamente para repetir um **BLOCO DE INSTRUÇÕES** (conjunto de linhas de códigos) sucessivas vezes

até atingir uma determinada **CONDIÇÃO** **(Lógica)** de parada.

$\longmapsto$ **BLOCO DE INSTRUÇÕES** (conjunto de linhas de códigos): Na linguagem de programação Python, o bloco é determinado por **espaços em branco razão de múltiplos de 4: 4, 8, 12, 16, 20, ....;** para cada novo bloco de controle.

```
while True:
  print('MENU DE OPÇÕES PARA O USUÁRIO:')
  print('OPÇÃO 1: Calcular a Equação do 2° Grau.')
  print('OPÇÃO 0: Sair do Programa.')
  opcao = int(input('Qual é a sua Opção: '))
  if opcao == 0:
    print('FIM DO PROGRAMA! TCHAU!!!')
    break
  elif opcao != 1:
    print('Opção incorreta do Menu. Escolha de novo, por favor!!!')
  else:
    A = float(input('Coeficiente A: '))
    B = float(input('Coeficiente B: '))
    C = float(input('Coeficiente C: '))
    if A == 0 or (B == 0 and C == 0):
      print('Não é uma Equação do 2° Grau.')
    else:
      delta = B ** 2 - 4 * A * C
      print(f'Valor de Delta: {delta: .2f}')
      if delta < 0:
        print('Não tem Raízes no conjunto dos números reais.')
      elif delta == 0:
        X1 = -B / (2 * A)
        print(f'Raiz X1 = X2 = {x1: .2f}')
      else:
        raizDelta = delta ** 0.5
        print(f'Valor da Raíz de Delta: {raizDelta: .2f}')
        X1 = (-B + raizDelta) / (2 * A)
        X2 = (-B - raizDelta) / (2 * A)
        print(f'Raiz X1 = {X1: .2f}')
        print(f'Raiz X2 = {X2: .2f}')
```

```
"""
EXEMPLO 4:
Enumerar e exibir na tela todas as sugestões de senhas numéricas
de 4 digitos com a seguinte "criptografia fraca":

-> ÍMPARES e (AND) MÚLTIPLAS DE 39.

VALOR INICIAL.           : 1000
VALOR FINAL (não incluso): 9999
"""
counter = 0
for i in range(1000,9999):
  if i % 39 == 0 and i % 2 != 0:
    counter += 1
    print(f'{counter}º NÚMERO: {i}')
```

```
"""
EXEMPLO 5:
Enumerar e exibir na tela os números ímpares entre
[X, X ** 3]. Onde X é um valor >= 1 lido do usuário.

VALOR INICIAL: X
VALOR FINAL: X ** 3
OBS: X ** 3 + 1, pois o VALOR FINAL: não incluso.
"""
try:
  x = int(input('Insira o número: '))
  print()
  if x >= 1:
    counter = 0
    print(f'Entre {x} e {x ** 3}, temos: ')
    for i in range(x, x ** 3 + 1):
      if i % 2 != 0:
        counter += 1
        print(f'{counter}º NÚMERO: {i}.')
  else:
    print('ERRO: Valor não é >= 1.')
except:
  print('ERRO: Você digitou algum valor incorretamente.')
```

```
"""
EXEMPLO 6:
Enumerar e exibir na tela os números ímpares E não múltiplos de 5 entre
[X + Y, X ** 2 + Y ** 2]. Onde X e Y são valores >= 1 lido do usuário.
Exibir na tela também a soma e a média deste números.

VALOR INICIAL: X + Y
VALOR FINAL. : X ** 2 + Y ** 2

OBS: (X ** 2 + Y ** 2) + 1, pois o VALOR FINAL: não incluso.
"""
try:
  x = int(input('Insira o valor de X: '))
  y = int(input('Insira o valor de Y: '))
  if x < 1 or y < 1:
    print('ERRO: Valores precisam ser maiores que 0.')
  else:
    counter = 0
    soma = 0
    for i in range(x + y, (x ** 2 + y ** 2) + 1):
      if i % 5 != 0 and i % 2 != 0:
        counter += 1
        soma += i
        print(f'{counter}º NÚMERO: {i}.')
    if counter > 0:
      media = soma / counter
      print()
      print(f'A soma dos valores deu {soma} e a média de todos deu {media:.2f}')
    else:
      print('\nNenhum número encontrado que atenda aos critérios.')
except:
  print('ERRO: Algum valor foi inserido incorretamente.')
```

```
"""
  EXEMPLO 7:
  Enumerar e exibir na tela todas as sugestões de senhas numéricas
  de 4 digitos (DISTINTOS ENTRE SI: NÃO REPETIDOS) com a seguinte "criptografia fraca":

  -> ÍMPARES e (AND) MÚLTIPLAS DE 39.

  VALOR INICIAL.           : 1000
  VALOR FINAL (não incluso): 9999
"""
counter = 0
for i in range(1000, 9999):
  if i % 2 != 0 and i % 39 == 0:
    d1 = i // 1000
    d2 = (i // 100) % 10
    d3 = (i // 10) % 10
    d4 = i % 10
    if len({d1, d2, d3, d4})== 4:
      counter += 1
      print(f'{counter}ª SENHA: {i}')
```

```
"""
EXEMPLO 1:
Ler a nota[0, 10] de x alunos (CONDIÇÃO: VALOR FINAL) de uma turma
e exibir sua a média geral.
Média = Soma as Notas de todos / Alunos
"""
counter = 0
soma = 0
try:
  x = int(input('Insira o total de alunos: '))
  if x <=2:
    print('ERRO: Valor é menor que 1')
  while counter < x:
    try:
      print(f'{counter + 1}º aluno, insira sua nota: ', end = '')
      nota = float(input())
      if nota < 0 or nota > 10:
        print('ERRO: Você não inseriu a nota no intervalo correto.')
      else:
        counter += 1
        soma = soma + nota
    except:
      print('ERRO: Insira o valor novamente.')
  media = soma / x
  print()
  print(f'A média dos {x} alunos deu {media:.2f}.')
except:
  print('ERRO: Você inseriu algum valor incorretamente.')
```

```
"""
EXEMPLO 2:
Ler a opinião de x alunos universitários (CONDIÇÃO: VALOR FINAL) sobre a seguinte questão:

Você gostou das aulas tele-presenciais? 1: SIM ou 0: NÃO

Exiba o resultado da pesquisa em porcentagem(%)

-> Como calcular a Porcentagem?
% Parte = Parte / Total * 100
% sim = sim / (sim + nao) * 100
% nao = nao / (sim + nao) * 100
"""
alunos = 0
sim, nao = 0, 0
try:
  x = int(input('Insira a quantidade de alunos: '))
  if x < 1:
    print('ERRO: Valor inserido é menor que 1.')
  while alunos < x:
    try:
      print(f'{alunos + 1}º Aluno, você gostou das aulas tele-presenciais? 1: SIM ou 2: NÃO: ', end = '')
      q = int(input())
      if q != 1 and q != 2:
        print('ERRO: Valor digitado incorretamente.')
      else:
        if q == 1:
          sim += 1
        else:
          nao += 1
        alunos += 1
        psim = sim / alunos * 100
        pnao = nao / alunos * 100
        print(f'Votos SIM = {sim} ({psim:.2f}%)')
        print(f'Votos NÃO = {nao} ({pnao:.2f}%)')
    except:
      print('ERRO: Insira o valor novamente.')
except:
  print('ERRO: Você inseriu algum valor incorretamente.')
```

## **Estrutura de Repetição com interrupção (comando break)**: Comando WHILE + BREAK.

# $\longmapsto$ **COMANDO WHILE(True) + BREAK**:

É mais adequado para problemas onde não é informado o VALOR FINAL.

while (True):
  
$\quad$ // AÇÃO 1

$\quad$  // AÇÃO 2

$\quad$  // AÇÃO 3

 $\quad$ // AÇÃO ...

$\quad$if (CONDIÇÃO DE PARADA == PARAR):

$\quad$$\quad$break

```
"""
EXEMPLO 2: WHILE + BREAK
Ler a opinião de VÁRIOS alunos universitários sobre a seguinte questão:

Você gostou das aulas tele-presenciais? 1: SIM ou 2: NÃO

Exiba o resultado da pesquisa em porcentagem(%)

CONDIÇÃO DE PARADA: Pare seu programa quando o usuário responder ZERO.

-> Como calcular a Porcentagem?
% Partes = Parte / Total * 100
% sim = sim / (sim + nao) * 100
% nao = nao / (sim + nao) * 100
"""
alunos = 0
sim, nao = 0, 0
try:
  while True:
    try:
      print(f'{alunos + 1}º Aluno, você gostou das aulas tele-presenciais? 1: SIM ou 2: NÃO: ou 0: PARAR ', end = '')
      q = int(input())
      if q < 0 or q > 2:
        print('ERRO: Valor digitado incorretamente.')
      else:
        if q == 1:
          sim += 1
        elif q == 0:
          print('FIM DA CONTAGEM DE VOTOS.')
          break
        else:
          nao += 1
        alunos += 1
        psim = sim / alunos * 100
        pnao = nao / alunos * 100
        print(f'Votos SIM = {sim} ({psim:.2f}%)')
        print(f'Votos NÃO = {nao} ({pnao:.2f}%)')
    except:
      print('ERRO: Insira o valor novamente.')
except:
  print('ERRO: Você inseriu algum valor incorretamente.')
```

```
'''
Escrever um algoritmo em python que leia VÁRIOS números reais (float) e exiba na tela:

-> Média dos números Positivos = Soma de números positivos / Quantidade de números positivos
-> Média dos números Negativos = Soma de números negativos / Quantidade de números negativos

Parar (BREAK) o programa quando o usuário digitar ZERO.
'''
counter = 0
numero, npositivo, nnegativo = 0, 0 , 0
positivo, negativo = 0, 0
while True:
  try:
    print(f'{counter + 1}º número, insira: ', end ='')
    numero = float(input())
    if numero < 0:
      nnegativo += numero
      negativo += 1
      nnegativo = nnegativo / negativo
      print(f'NÚMEROS NEGATIVOS = {negativo}, MÉDIA DOS NÚMEROS NEGATIVOS: {nnegativo:.2f}')
    elif numero > 0:
      npositivo += numero
      positivo += 1
      npositivo = npositivo / positivo
      print(f'NÚMEROS POSITIVOS = {positivo}, MÉDIA DOS NÚMEROS POSITIVOS: {npositivo:.2f}')
    else:
      print('FIM DO PROGRAMA.')
      break
    counter += 1
  except:
    print('ERRO: Valor incorreto, digite novamente.')
```
