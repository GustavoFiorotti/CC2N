# EXERCÍCIO NÚMERO 2:

```
quantidade = 0
soma = 0
media = 0
for contador in range(200, 100, -1):
  if contador % 11:
    quantidade += 1
    soma = contador + quantidade
    media = soma / quantidade
    print(f'{quantidade}° NÚMERO: {contador}.')
print(f'Soma de todos os múltiplos é {soma} e a média é {media:.2f}.')
```

# EXERCÍCIO NÚMERO 3:

```
# TRATANMENTO DE EXCEÇÃO
try:
  Li = int(input('ENTRE COM O LIMITE INFERIOR (>= 1) DO INTERVALO: '))
  Ls = int(input('ENTRE COM O LIMITE SUPERIOR (>= 1) DO INTERVALO: '))
  N = int(input('ENTRE COM O MÚLTIPLO DESEJADO (>= 2): '))
  # TRATAMENTO DE ERRO:
  if(Li < 1 or Ls < 1 or Ls < Li):
    print('ERRO: Dados de entrada do problema')
  elif N < 1:
    print('MÚLTIPLO MENOR QUE 1.')
  else:
    quantidade = 0
    for contador in range(Li, Ls + 1, N):
      quantidade += 1
      print(f'{quantidade}° NÚMERO: {contador}.')
except:
  print('ERRO: número inserido incorreto.')
```

# EXERCÍCIO NÚMERO 4:

```
try:
  quantidade = 0
  choose = int(input('Entre 10 e 99, qual múltiplo você quer primeiro, PARES(1) ou ÍMPARES(2)?: '))
  if choose == 1:
    for contador in range (10, 100, 2):
      quantidade += 1
      print(f'{quantidade}º NÚMERO: {contador}.')
    print()
    print('Agora, os ímpares:')
    for contador in range (11, 100, 2):
      quantidade += 1
      print(f'{quantidade}º NÚMERO: {contador}.')

  elif choose == 2:
    for contador in range (11, 100, 2):
      quantidade += 1
      print(f'{quantidade}º NÚMERO: {contador}.')
    print()
    print('Agora, os pares:')
    for contador in range (10, 100, 2):
      quantidade += 1
      print(f'{quantidade}º NÚMERO: {contador}.')
  else:
    print('Você não inseriu nem 1, nem 2.')
except:
  print('ERRO: Você inseriu o dado incorretamente.')
```
