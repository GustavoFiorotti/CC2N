```
try:
    sexo = input('Você é homem (M) ou mulher (F)?: ').upper()

    if sexo == "M" or sexo == "F":
        h = float(input('Insira a sua altura: '))

        if sexo == "M":
            massa = (72.7 * h) - 58
        else:
            massa = (62.1 * h) - 44.7

        print(f'O seu peso é {massa:.2f} Kgs.')
    else:
        print('ERRO: Insira M para homem ou F para mulher.')

except ValueError:
    print('ERRO: Insira um valor numérico válido para a altura.')
```

```
try:
    medida = float(input('Insira o que você quer, Celsius (1) ou Fahrenheit (2): '))

    if medida == 1:
        C = float(input('Insira o valor da temperatura em Celsius: '))

        F = (C * 9/5) + 32

        print()
        print(f'{C} graus Celsius em Fahrenheit é {F:.1f} graus Fahrenheit.')

    elif medida == 2:
        F = float(input('Insira o valor da temperatura em Fahrenheit: '))

        C = (F - 32) * 5 / 9

        print()
        print(f'{F} graus Fahrenheit em Celsius é {C:.1f} graus Celsius.')

    else:
        print()
        print('ERRO: O valor digitado não é 1 OU 2.')
except ValueError:
    print('ERRO: O valor digitado não é um número.')
```

```
try:

  v1 = int(input('Insira o PRIMEIRO valor: '))
  v2 = int(input('Insira o SEGUNDO valor: '))
  v3 = int(input('Insira o TERCEIRO valor: '))

  if v1 == v2 or v1 == v3 or v2 == v3:
    print('Valores iguais.')

  else:

    print()

    if v1 > v2:

      soma = (v2 + v3) /2

      print(f'A soma dos maiores irá dar {soma}.')

    elif v2 > v3:

      soma = (v2 + v1) /2

      print(f'A soma dos maiores irá dar {soma}.')

    elif v3 > v2:

      soma = (v1 + v3) /2

      print(f'A soma dos maiores irá dar {soma}.')

except:
  print('ERRO: Valores incorretos.')
```
