```
def funcao(n,choose):
  try:
    mult = 1
    for i in range(1,n + 1):
      if choose == 1:
        if i % 2 == 0:
          mult *= i
      elif choose == 2:
        if i % 2 != 0:
          mult *= i
    return mult
  except:
    return
counter = 0
while counter < 5:
  try:
    print(f'{counter + 1}º PESSOA, Digite o número: ', end='')
    n = int(input())
    if n < 1:
      print('ERRO: Digite novamente.')
    else:
      choose = int(input('Par(1) ou ímpar(2)?: '))
      if choose < 1 or choose > 2:
        print('ERRO: Você inseriu um valor fora da variável.')
      else:
        mult = funcao(n, choose)
        print(f'Multiplicação dos valores deu: {mult}.')
        counter += 1
    print()
  except:
    print('ERRO: Digite novamente.')
```

```
def fat(x):
  if x == 0 or x == 1:
    return 1
  else:
    resultado = x * fat(x - 1)
    return resultado
def arran(n, p):
  arranjo = fat(n) / fat(n-p)
  return arranjo
def comb(n,p):
  combinacao = arran(n,p) / fat(p)
  return combinacao

counter = 0
while True:
  print(f'{counter + 1}º usuário, Digite (qualquer valor) ou (0) para finalizar o programa: ',end='')
  choose = int(input())
  if choose == 0:
    print('PROGRAMA ENCERRADO.')
    break
  else:
    p = int(input('digite o valor de P: '))
    if p == '00':
      print('PROGRAMA FINALIZADO.')
      break
    else:
      try:
        n = int(input('Insira o valor de N: '))
        if p < 0 or n < 0:
          print('ERRO: Valor menor que 0, digite novamente.')
        elif n < p:
          print('ERRO: Valor de P é maior que N.')
        else:
          counter += 1
          arranjo = arran(n,p)
          combinacao = comb(n,p)
          print()
          print(f'Valor do arranjo = {arranjo:.2f}.')
          print(f'Valor da combinação = {combinacao:.2f}.')
          print()
      except:
        print('ERRO: Digite novamente.')
```

```
def fibonacci(n):
  if n == 1 or n == 2:
    return 1
  elif n == 0:
    return 0
  else:
    return fibonacci(n-1) + fibonacci(n-2)
def soma(n):
  soma = 0
  for i in range(n + 1):
    soma += fibonacci(i)
  return soma
try:
  n = int(input('Quantos primeiros termos de Fibonacci você quer contar?: '))
  if n < 0:
    print('ERRO: Valor é menor que 0.')
  else:
    calcular = soma(n)
    print()
    print(f'Soma dos {n} primeiros termos de Fibonacci deu {calcular}.')
except:
  print('ERRO: Digite novamente.')
```

```
import math
def calcular_a(n, k):
  termos_a = []
  for i in range(n):
    termo = k / (math.factorial(i) + k)
    termos_a.append(termo)
  return termos_a

def calcular_b(n, k, p):
  termos_b = []
  for i in range(1, n + 1):
    termo = (k**i - k) / (p * i)
    termos_b.append(termo)
  return termos_b

k = math.pi
p = math.e
counter = 0
while True:
  try:
    print(f'{counter + 1}º USUÁRIO, INSIRA (1) PARA CONTINUAR ou (0) PARA ENCERRAR O PROGRAMA: ', end='')
    escolha = int(input())
    if escolha == 0:
      print('PROGRAMA ENCERRADO.')
      break
    elif escolha != 1:
      print('ERRO: Opção inválida. Tente novamente.')
    else:
      n = int(input('INSIRA O NÚMERO DE TERMOS (n): '))
      if n <= 0:
        print('ERRO: O número de termos deve ser um inteiro positivo.')
      else:
        termos_a = calcular_a(n, k)
        print(f'TERMOS DA FUNÇÃO A: {termos_a}')
        print(f'SOMA DOS TERMOS DA FUNÇÃO A: {sum(termos_a)}')
        print(f'MÉDIA DOS TERMOS DA FUNÇÃO A: {sum(termos_a) / n}')
        termos_b = calcular_b(n, k, p)
        print(f'TERMOS DA FUNÇÃO B: {termos_b}')
        print(f'SOMA DOS TERMOS DA FUNÇÃO B: {sum(termos_b)}')
        print(f'MÉDIA DOS TERMOS DA FUNÇÃO B: {sum(termos_b) / n}')
        counter += 1
        print()
  except:
    print('ERRO: Digite novamente um valor válido.')
```
