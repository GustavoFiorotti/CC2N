# EXEMPLO 1: Distância entre os pontos cartesianos: $\ P$ e $\ Q$.

Distancia entre $ \ P(Px, Py)$ e $ \ Q(Qx, Qy)$ = $ \sqrt{(P_x - Q_x)^2 + (P_y - Q_y)^2}$

```
def distanciaEntrePontos(Px = 0, Py = 0, Qx = 0, Qy = 0):
  """
  HELP - DOCSTRING: Cálculo entre os pontos cartesianos:
  P(Px, Py) e Q(Qx, Qy).
  Por default (padrão): P(0, 0) e Q(0, 0).
  """
  distancia = ((Px - Qx) ** 2 + (Py - Qy) ** 2) ** 0.5
  return distancia
```

# EXEMPLO 4: Função para estabelecer os índices de Saúde.

### CÉLULA 1: Declaração da Função: Índices de Saúde.

```
def RelatorioSaude(massa, altura, medidaCintura, medidaQuadril):
  """
    AJUDA DA FUNÇÃO (DOCSTRING):
    Parâmetros de entrada:
    massa: em quilogramas;
    altura: em quilogramas;
    medidaCintura: em centímetros;
    medidaQuadril: em centímetros;

    Índices de Saúde implementa como objetivo determinar os parâmetros de
    referência de saúde com os seguintes indicadores, a saber:
    -> RCQ: Relação Cintura-Quadril = medidaCintura / medidaQuadril * 100 (%)
    -> IMC: Índice de Massa Corpórea = Massa / (altura ** 2)
    -> Classificação por IMC, sendo:
        <> "MAGREZA": IMC < 18.5
        <> "PESO NORMAL": IMC < 25
        <> "SOBREPESO": IMC < 30
        <> "OBESIDADE GRAU I": IMC < 35
        <> "OBESIDADE GRAU II": IMC < 40
        <> "OBESIDADE GRAU III": IMC >= 40
  """
  try:
    RCQ = medidaCintura / medidaQuadril * 100
    IMC = massa / (altura ** 2)
    if(IMC < 18.5):
      classificacao = 'MAGREZA'
    elif(IMC < 25.0):
      classificacao = 'PESO NORMAL'
    elif(IMC < 30.0):
      classificacao = 'SOBREPESO'
    elif(IMC < 35.0):
      classificacao = 'OBESIDADE GRAU I'
    elif(IMC < 40.0):
      classificacao = 'OBESIDADE GRAU II'
    else:
      classificacao = 'OBESIDADE GRAU III'
    return RCQ, IMC, classificacao
  except:
    return
```

```
# Lógica do while + break: Menu de Opções para o Médico
pacientes = 0
while True:
  try:
    print('MENU:')
    print('1: INSERIR PACIENTE')
    print('0: SAIR DO PROGRAMA')
    opcao = int(input('SUA OPÇÃO: '))
    if(opcao == 0):
      print('FIM DO PROGRAMA')
      break
    elif(opcao != 1):
      print('ERRO: OPÇÃO INCORRETA. ESCOLHA DE NOVO.')
    else:
      print(f'DADOS DO {pacientes + 1}º PACIENTE: ')
      # INPUT: MASSA
      while True:
        # TRATAMENTO DE EXCEÇÃO
        try:
          massa = float(input('SUA MASSA (em quilogramas): '))
          # TRATAMENTO DE ERRO:
          if(massa <= 0):
            print('ERRO: DADOS DE ENTRADA')
          else:
            break
        except:
          print('ERRO: DADOS DE ENTRADA')
      # INPUT: ALTURA
      while True:
        # TRATAMENTO DE EXCEÇÃO
        try:
          altura = float(input('SUA ALTURA (em metros): '))
          # TRATAMENTO DE ERRO:
          if(altura <= 0):
            print('ERRO: DADOS DE ENTRADA')
          else:
            break
        except:
          print('ERRO: DADOS DE ENTRADA')
      # INPUT: MEDIDA DA CINTURA
      while True:
        # TRATAMENTO DE EXCEÇÃO
        try:
          medidaCintura = float(input('Medida da Cintura (em Centimetros): '))
          # TRATAMENTO DE ERRO:
          if(medidaCintura <= 0):
            print('ERRO: DADOS DE ENTRADA')
          else:
            break
        except:
          print('ERRO: DADOS DE ENTRADA')
      # INPUT: MEDIDA DO QUADRIL
      while True:
        # TRATAMENTO DE EXCEÇÃO
        try:
          medidaQuadril = float(input('Medida do Quadril (em Centimetros): '))
          # TRATAMENTO DE ERRO:
          if(medidaQuadril <= 0):
            print('ERRO: DADOS DE ENTRADA')
          else:
            break
        except:
          print('ERRO: DADOS DE ENTRADA')
      print('RELATÓRIO DE SAÚDE:')
      RCQ, IMC, classificacao = RelatorioSaude(massa, altura, medidaCintura, medidaQuadril)
      print(f'RCQ: {RCQ: .1f}')
      print(f'IMC: {IMC: .1f}')
      print(f'CLASSIFICAÇÃO: {classificacao}')
      pacientes += 1
  except:
    print('ERRO: OPÇÃO INCORRETA. ESCOLHA DE NOVO.')
```

```
def numeroPerfeito(numero = 1):
  """
    DOCSTRING: MODULARIZAR + ENCAPSULAR
    Na Matemática, um número perfeito é um número natural
    para o qual a soma de todos os seus divisores naturais
    próprios (excluindo ele mesmo) é igual ao próprio número.

    Por exemplo, o número 6 e 28 são , pois:
    6  = 1 + 2 + 3
    28 = 1 + 2 + 4 + 7 + 14
    ...
    Numero = 1 + 2 (START) + 3 + ... + Numero / 2 (STOP)
  """
  try:
    soma = 1
    for contador in range(2, numero // 2 + 1):
      if(numero % contador == 0):
        soma = soma + contador
    if(soma == numero):
      return True
    else:
      return False
  except Exception as erro:
    return f'ERRO: EXCEÇÃO {erro}'
```

```
def fat(x):
  if x == 0 or x == 1:
    return 1
  else:
    resultado = x * fat(x - 1)
    return resultado
def arran(n, p):
  arranjo = fat(n) / fat(n-p)
  return arranjo
def comb(n,p):
  combinacao = arran(n,p) / fat(p)
  return combinacao

counter = 0
while True:
  print(f'{counter + 1}º usuário, Digite (qualquer valor) ou (0) para finalizar o programa: ',end='')
  choose = int(input())
  if choose == 0:
    print('PROGRAMA ENCERRADO.')
    break
  else:
    p = int(input('digite o valor de P: '))
    if p == '00':
      print('PROGRAMA FINALIZADO.')
      break
    else:
      try:
        n = int(input('Insira o valor de N: '))
        if p < 0 or n < 0:
          print('ERRO: Valor menor que 0, digite novamente.')
        elif n < p:
          print('ERRO: Valor de P é maior que N.')
        else:
          counter += 1
          arranjo = arran(n,p)
          combinacao = comb(n,p)
          print()
          print(f'Valor do arranjo = {arranjo:.2f}.')
          print(f'Valor da combinação = {combinacao:.2f}.')
          print()
      except:
        print('ERRO: Digite novamente.')
```

```
def calcular(a,b,c):
  if (a < b + c) and (b < a + c) and (c < a + b):
    if len({a,b,c}) == 3:
      return 'ESCALENO'
    elif len({a,b,c}) == 2:
      return 'ISÓCELES'
    else:
      return 'EQUILÁTERO'
  else:
    return 0
counter = 0
while True:
  try:
    print(f'{counter + 1}º USUÁRIO, INSIRA (QUALQUER VALOR) PARA INICIAR O PROGRAMA ou (0) PARA ENCERRAR: ',end='')
    choose = int(input())
    if choose == 0:
      print('PROGRAMA ENCERRADO.')
      break
    else:
      a = float(input('INSIRA O VALOR DE A: '))
      if a == '00':
        print()
        print('PROGRAMA ENCERRADO.')
        break
      if a < 0:
        print('ERRO: Valor menor que 0.')
      else:
        try:
          b = float(input('INSIRA O VALOR DE B: '))
          if b < 0:
            print('ERRO: Valor menor que 0.')
          else:
            c = float(input('INSIRA O VALOR DE C: '))
            if c < 0:
              print('ERRO: Valor menor que 0.')
            else:
              counter += 1
              calcular = calcular(a,b,c)
              print()
              if calcular:
                print(f'O TRIÂNGULO É UM TRIÂNGULO {calcular}.')
              else:
                print('NÃO É UM TRIÂNGULO.')
        except:
          print('ERRO: Digite novamente.')
  except:
    print('ERRO: Digite novamente.')
```

```
import math
def calcular_a(n, k):
  termos_a = []
  for i in range(n):
    termo = k / (math.factorial(i) + k)
    termos_a.append(termo)
  return termos_a

def calcular_b(n, k, p):
  termos_b = []
  for i in range(1, n + 1):
    termo = (k**i - k) / (p * i)
    termos_b.append(termo)
  return termos_b

k = math.pi
p = math.e
counter = 0
while True:
  try:
    print(f'{counter + 1}º USUÁRIO, INSIRA (1) PARA CONTINUAR ou (0) PARA ENCERRAR O PROGRAMA: ', end='')
    escolha = int(input())
    if escolha == 0:
      print('PROGRAMA ENCERRADO.')
      break
    elif escolha != 1:
      print('ERRO: Opção inválida. Tente novamente.')
    else:
      n = int(input('INSIRA O NÚMERO DE TERMOS (n): '))
      if n <= 0:
        print('ERRO: O número de termos deve ser um inteiro positivo.')
      else:
        termos_a = calcular_a(n, k)
        print(f'TERMOS DA FUNÇÃO A: {termos_a}')
        print(f'SOMA DOS TERMOS DA FUNÇÃO A: {sum(termos_a)}')
        print(f'MÉDIA DOS TERMOS DA FUNÇÃO A: {sum(termos_a) / n}')
        termos_b = calcular_b(n, k, p)
        print(f'TERMOS DA FUNÇÃO B: {termos_b}')
        print(f'SOMA DOS TERMOS DA FUNÇÃO B: {sum(termos_b)}')
        print(f'MÉDIA DOS TERMOS DA FUNÇÃO B: {sum(termos_b) / n}')
        counter += 1
        print()
  except:
    print('ERRO: Digite novamente um valor válido.')
```
